Use matplotlib to plot the following equation:
y = x2 - x + 2
You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, the origin. -->

```python

import matplotlib.pyplot as plt
def num_generator(lst_of_x):
    x_num_list = []
    y_num_list= []
    for x in lst_of_x:
        y = (x **2 - x + 2)
        y_num_list.append(y)
    return (y_num_list)
num_generator([1,2,3])

x = list(range(-100, 101, 1))#[0,1,2,3,4,5,6,7]
y = num_generator(x)
y 
plt.figure(figsize =(10,10))
plt.plot(x,y, color = "red")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("X vs Y : My first plot")
plt.annotate("Origin", xy = (0,0), xytext = (0,2000), arrowprops = {"arrowstyle": "simple", 'facecolor': 'blue'})
plt.axhline(y = 0, color = 'black', ls= ':')
plt.axvline(color='black', ls = ':')
plt.show()
```

```python
x = list(range(-10,10))
m = []
for val in x:
    z = val**3
    m.append(z)
m
plt.figure(figsize = (10,10))

plt.plot(x,m, color = "hotpink", ls =":")
plt.title("$y = x^3$")
plt.annotate("Origin", (0,0), xytext = (2.5, -250), arrowprops = {'arrowstyle': 'simple', 'facecolor': 'blue'})
plt.axhline(y = 0, color = 'black', ls= ':')
plt.axvline(color='black', ls = ':')
plt.show()
```

```python
x = list(range(-10,10))
n = []
for val in x:
    z = 2**val
    n.append(z)
n
plt.figure(figsize = (10,10))
plt.plot(x,n, color = "red")
plt.xlabel(" x values")
plt.ylabel("y values")
plt.title("$y = 2^x$")
plt.annotate("Origin", (0,0), xytext=(-5,100), arrowprops = {'arrowstyle': 'wedge', 'facecolor': 'purple'})
plt.axhline(y = 0, color = 'black', ls= ':')
plt.axvline(color='black', ls = ':')
plt.show()
```

3. Combine the figures you created in the last step into one large figure with 4 subplots.


```python

x = list(range(0,50))
y0 = [(val**2-val+2) for val in x]
y1 = [val**(1/2) for val in x]
y2 = [val **3 for val in x]
y3 = [2**val for val in x]

plt.figure(figsize = (12,12))
plt.subplot(2,2,1)
plt.plot(x,y0, color = "purple")
plt.title("$y = x^2 - x +2$")

plt.subplot(2,2,2)
plt.plot(x, y1, color = "red")
plt.title(" $y = sqrt(x)$")

plt.subplot(2,2,3)
plt.scatter(x,y2, color = "green")
plt.title(" $y = x ^ 3$")

plt.subplot(2,2,4)
plt.bar(x,y3, color = "purple")
plt.title("$y = 2^x$")
plt.axhline(y = 0, color = 'black', ls= ':')
plt.axvline(color='black', ls = ':')
plt.show()
```
