```python
#3. Would it be reasonable to predict species based on sepal width and sepal length?

# lets try to answer this question by first getting the area of the sepal and the area of the petal

sepal_area = iris.sepal_length * iris.sepal_width
petal_area = iris.petal_length * iris.petal_width
iris.insert(2, 'sepal_area', sepal_area)
sns.lmplot(x= 'sepal_width', y = "sepal_length", hue ='species', data = iris)
```

```python
#4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items
#and the revenue produced by each.

from env import host, user, password

url = f'mysql+pymysql://{user}:{password}@{host}/chipotle'

chipotle = pd.read_sql('''select * from orders''', url)
chipotle.head(10)

# converting the string item_price to float for easier mathematical operations

chipotle['item_price'] = chipotle['item_price'].str.replace('$', '').astype(float)

# grouping by item name and then getting the sum of the quantity and the price and selecting the top four

top_four= chipotle.groupby('item_name')[['quantity', 'item_price']].sum().nlargest(4, 'quantity', keep = 'all')

# plotting the bar plot using plt.bar from matplotlib

plt.bar(top_four.index, top_four.item_price, color = 'purple')
plt.xticks(rotation = 45)

```

```python
#5. Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the individual 
#subject's reaction times and a more prominant line showing the average change in reaction time.
data('sleepstudy', show_doc = True)
```

```python
sleepstudy = data('sleepstudy')
sleepstudy.describe()
```

```python

# calculating the average reaction time after groupin the data by days
avg = sleepstudy.groupby('Days')[['Reaction']].agg('mean')


plt.figure(figsize = (14,12)) # setting a bigger chart

# plotting the reaction time of the subject on different days
sns.lineplot(x = 'Days', y ='Reaction', hue = 'Subject', data = sleepstudy, palette = 'colorblind')

# plotting the average reaction time over 10 days 
plt.plot(avg.index, avg.Reaction, color = 'black', ls = ':', alpha = 1, linewidth = 6, label = "average reaction time")
plt.legend()
plt.annotate('Average Reaction Time',(6, 310), xytext= (2, 450), fontsize = 15,  arrowprops = {'facecolor': 'purple', 
                                                                              'arrowstyle': 'wedge'} )
```
